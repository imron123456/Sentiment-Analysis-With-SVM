```python
# üìö Import library
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
import pandas as pd

# üîß Load pretrained sentiment model
model_name = "mdhugol/indonesia-bert-sentiment-classification"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# üöÄ Setup pipeline
sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# üìÅ Upload file Excel berisi komentar
from google.colab import files
uploaded = files.upload()  # Seret atau pilih file .xlsx kamu di sini

# üìñ Baca file Excel
df = pd.read_excel(next(iter(uploaded)))  # Ambil nama file pertama yg diupload
print("Contoh data:", df.head())

# ‚úÖ Ubah nama kolom jika perlu (ganti 'komentar' sesuai nama kolom di file kamu)
text_column = "komen"

# üîç Prediksi sentimen
results = sentiment_pipeline(df[text_column].tolist())

# üè∑Ô∏è Label mapping
label_map = {'LABEL_0': 'positive', 'LABEL_1': 'neutral', 'LABEL_2': 'negative'}
df['sentiment_label'] = [label_map[result['label']] for result in results]
df['sentiment_score'] = [result['score'] for result in results]

# üíæ Simpan hasil ke file baru
output_filename = "hasil_sentimen.xlsx"
df.to_excel(output_filename, index=False)

# üì• Download file hasil
files.download(output_filename)
```
