```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
import ast  # untuk konversi string list jadi list Python

# ==============================
# Download stopwords NLTK
# ==============================
nltk.download('stopwords')

# ==============================
# Load file Excel
# ==============================
file_name = 'Data Pakar Tokenized.xlsx'  # Ganti sesuai nama file kamu
df = pd.read_excel(file_name)

# ==============================
# Buat daftar stopword
# ==============================
stop_words_id = set(stopwords.words('indonesian'))
stop_words_en = set(stopwords.words('english'))
stop_words = stop_words_id.union(stop_words_en)

# ==============================
# Fungsi untuk stopword removal
# ==============================
def remove_stopwords(token_list):
    # Jika token disimpan sebagai string list, konversi ke list Python dulu
    if isinstance(token_list, str):
        token_list = ast.literal_eval(token_list)
    filtered_tokens = [word for word in token_list if word.lower() not in stop_words]
    return filtered_tokens

# ==============================
# Terapkan stopword removal
# ==============================
df['stopword_clean'] = df['tokenized'].apply(remove_stopwords)

# ==============================
# Simpan hasil ke file baru
# ==============================
df.to_excel('Data Pakar Stopword.xlsx', index=False)

print('Stopword removal selesai! Hasil disimpan ke data_token_clean.xlsx')
```
