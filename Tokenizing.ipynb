```python
import pandas as pd
import spacy
import string
import sys

# Load spaCy model
nlp = spacy.load("en_core_web_sm")  # Ganti model jika perlu, misal untuk multi-bahasa

def clean_text(text):
    """Membersihkan teks dari spasi extra"""
    if not isinstance(text, str):
        return ""
    return text.strip().lower()

def spacy_tokenizer(text):
    text = clean_text(text)
    doc = nlp(text)
    # Ambil hanya token kata, buang tanda baca & spasi kosong
    tokens = [token.text for token in doc if not token.is_punct and not token.is_space]
    return tokens

def tokenize_column(input_file, output_file, column_name):
    try:
        # Baca file Excel
        df = pd.read_excel(input_file)

        if column_name not in df.columns:
            print(f"Kolom '{column_name}' tidak ditemukan di file.")
            sys.exit(1)

        # Tokenizing pakai spaCy
        df['tokenized'] = df[column_name].apply(spacy_tokenizer)

        # Simpan ke file baru
        df.to_excel(output_file, index=False)
        print(f"Proses selesai! File disimpan ke '{output_file}'")

    except FileNotFoundError:
        print(f"File '{input_file}' tidak ditemukan.")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")

if __name__ == "__main__":
    # Ganti sesuai kebutuhan
    input_file = "Data Pakar Cleaning.xlsx"
    output_file = "Data Pakar Tokenized.xlsx"
    column_name = "komen"  # Ganti dengan nama kolom di Excel Anda

    tokenize_column(input_file, output_file, column_name)
```
